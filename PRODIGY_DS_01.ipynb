```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
df = pd.read_csv('/Users/mostaphaatta/Downloads/Internship/classification.csv')

data = {
    'Category' : df['DN'].unique(),
    'Count' : df['DN'].value_counts()
}

colors = ['darkgoldenrod','forestgreen','grey','dodgerblue']
```

```python
style.use('classic')

plt.figure(figsize = (5,5),dpi=100, num='test', facecolor = 'w', edgecolor = 'w')
plt.bar(data['Category'], data['Count'], color = colors, width = 0.5)

plt.xlabel('land type')
plt.ylabel('pixel frequency')
plt.title('land classification')

plt.xticks(ticks=data['Category'], labels=['Soil','Vegetation','Built-up','Water'])

plt.savefig("land_class_dist.png")
```
